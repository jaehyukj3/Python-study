## Pandas를 활용한 데이터 전처리

```python
tips1['time'].unique()
```

```python
# 데이터 치환
# 사본의 값을 변경하여도 원본 tips 데이터는 변하지 않음
tips1.loc[0,'time'] = 'Lunch'
tips1.head()
```

```python
# 새 컬럼 추가하여 데이터 만들기(논리검색을 사용)
# .loc를 사용하여 논리검색을 행 칸에 넣고 열 칸에 새로운 컬럼 이름을 넣음, 데이터는 오른쪽에 넣음
tips.loc[tips['tip']>3.6, 'level'] = 'good'
tips.loc[(tips['tip']<=3.6)&(tips['tip'] >= 2), 'level'] = 'soso'
tips.loc[tips['tip']<2, 'level'] = 'bad'
print(tips['level'].isnull().sum())
tips
```

```python
tips.loc[tips['level']!='good',['tip','level']].head()
```

```python
# 새 컬럼 추가하여 데이터 만들기(반복문)
rows = tips.shape[0]
print(rows)
for i in range(rows):
    tips.loc[i,'index_id'] = str(i * 2)# str은 문자열로 변환, 예 1 --> '1'
tips
```

```python
tips_m = tips.set_index('index_id')
tips_m.head()
```

```python
tips_m
```

```python
# loc는 index 이름을 넣어줘야 함
tips_m.loc['2']
```

```python
# iloc는 index 이름과 상관 없이 행의 고유 넘버를 넣어줘야 함
tips_m.iloc[1]
```

```python
tips.head()
```

```python
# iloc은 마지막 값이 미포함
tips.iloc[:-1]
```

```python
# loc은 마지막 값이 포함
tips.loc[:1]
```

```python
# ratio 컬럼은 tip을 total_bill로 나눈 비율을 데이터로 함
tips['ratio'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
# 함수를 사용해서 새로운 컬럼 생성
# bill = totla_bill - tip
def bill(x):
    total_bill = x['total_bill']
    tip = x['tip']
    bill = total_bill - tip
    return bill

tips['bill'] = tips.apply(bill, axis='columns')
tips
```

```python
# one-hot encoding과 빈 값 채우는 방법
# 논리 검색, loc 를 사용해서 새 컬럼 생성
tips.loc[tips['level']=='good', 'level_good']=1
tips.loc[tips['level']=='soso', 'level_soso']=1
tips.loc[tips['level']=='bad', 'level_bad']=1
tips
```

```python
tips[tips['level_good'].notnull()]
```

```python
#빈 값 채우는 법
tips['level_good'] = tips['level_good'].fillna(0)
tips['level_soso'] = tips['level_soso'].fillna(0)
tips['level_bad'] = tips['level_bad'].fillna(0)
tips
```

```python
# one-hot encoding 컬럼 생성 방법
tips['level_good'] = 0
tips['level_soso'] = 0
tips['level_bad'] = 0
tips.loc[tips['level']=='good', 'level_good']=1
tips.loc[tips['level']=='soso', 'level_soso']=1
tips.loc[tips['level']=='bad', 'level_bad']=1
tips
```

```python
tips.describe()
```

```python
# &와 | 활용법
tips.loc[(tips['total_bill']>24)&(tips['tip']>3.5), 'customer'] = 'vip'
tips.loc[(tips['total_bill']<=24)|(tips['tip']<=3.5), 'customer'] = 'non-vip'
tips
```

```python
tips['total_bill(pre)'] = tips['total_bill'].shift(+1)
tips['total_bill(next)'] = tips['total_bill'].shift(-1)
tips[['total_bill','total_bill(pre)','total_bill(next)']]
```

```python
tips
```

```python
#불필요한 컬럼 제거하기
# axis에서 1은 columns와 같고 0은 rows와 같음
tips_s = tips.drop(['index_id', 'total_bill(pre)', 'total_bill(next)'], axis='columns')
tips_s
```
